## 载入实际数据

```python
df = pd.read_csv('./ac_on_rate.csv', index_col='time', usecols=[1, 2, 3])
```

```python
df
```

## 总开机率

```python
df_max = df.apply(lambda x : x.max(), axis=1)
```

```python
df_max
```

### 数据格式化

```python
df_ac_on = df_max.round(2)
```

```python
df_ac_on
```

```python
df_ac_on_minutes = df_ac_on[df_ac_on>0]
```

### 计算开机分钟数

```python
df_ac_on_info = df_ac_on_minutes.describe()
```

```python
df_ac_on_info['count']
```

## 开机率min/max值

```python
df_all = df[df > 0]
```

```python
df_all = df_all.dropna(how='all')
```

```python
df_all
```

```python
df_all_info = df_all.describe()
```

```python
df_all_info
```

```python
ac_on_minutes = df_all_info.loc['count'].tolist()
```

```python
ac_on_minutes
```

```python
ac_on_min = min(ac_on_minutes)
```

```python
ac_on_max = max(ac_on_minutes)
```

```python
ac_on_min
```
